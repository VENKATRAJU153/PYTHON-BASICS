```python
#1.	Create an empty tuple.
ttuple = ()
print(ttuple)
```

```python
#2.	Create a tuple with 5 integers.
ttuple = (1,2,3,4,5)
print(ttuple)
```

```python
#3.	Access 3rd element of a tuple.
ttuple = (1,2,3,4,5)
print(ttuple[2])
```

```python
#4.	Access last element without negative index.
ttuple = (1,2,3,4,5)
x = len(ttuple)-1
print(ttuple[x])
```

```python
#5.	Slice first three elements.
ttuple = (1,2,3,4,5)
print(ttuple[:3])
```

```python
#6.	Slice every alternate value.
ttuple = (11,22,13,46,15,12,34,56,23,45,)
print(ttuple[: : 2])
```

```python
#7.	Reverse a tuple using slicing.
ttuple = (11,22,13,46,15,12,34,56,23,45,)
print(ttuple[: : -1])
```

```python
#8.	Concatenate two tuples.
t1= (11,22,13,46,15,12,34,56,23,45,)
t2 = ('a','b','c')
t3 = t1 + t2
print(t3)
```

```python
#9.	Replicate tuple (1,2) twice.
t =(1,2)
t * 2
```

```python
#10.	Find length of a tuple.
t1= (11,22,13,46,15,12,34,56,23,45,)
print(len(t1))
```

```python
#11.	Convert a list to a tuple.
l1= [11,22,13,46,15,12,34,56,23,45,]
t1 = tuple(l1)
print(type(t1))
print(t1)
```

```python
#12.	Convert a string "python" to tuple of characters.
st1 = "python"
t1 = tuple(st1)
print(t1)
```

```python
#13.	Check if value exists in tuple.
t1= (11,22,13,46,15,12,34,56,23,45,)
v = int(input("enter a value"))
if v in t1:
  print("EXIST")
else:
  print("NOT EXIST")
```

```python
#14.	Find maximum element in tuple.
t1= (11,22,13,46,15,12,34,56,23,45,)
print(max(t1))
```

```python
#15.	Find minimum element in tuple.
t1= (11,22,13,46,15,12,34,56,23,45,)
print(min(t1))
```

```python
#16.	Count how many times 2 appears in (1,2,2,3).
t1= (1,2,2,3,2,4,1,6,7,3)
t1.count(2)
```

```python
#17.	Find index of an element.
t1= (11,22,13,46,15,12,34,56,23,45,)
v = int(input("enter a value"))
if v in t1:
  print(t1.index(v))
else:
  ("value not exist")
```

```python
#18.	Perform tuple packing.
t1 = tuple(input())
print(t1)
```

```python
#t1= (11,22,13,46,15,12,34,56,23,45,)
t1= (11,22,13,46,15,12,34,56,23,45,)
(a,b,*c,d,e,f) = t1
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
```

```python
#1.	Write a program that prints numbers from 1 to 20 using a for loop.
num = int(input())
for i in range(1,num+1):
  print(i, end=" ")
```

```python
#2.	Write a program using a while loop to print even numbers between 1 and 50.
num = 1
while num<=50:
  if num%2==0:
    print(num, end=" ")
  num +=1
```

```python
#3.	Write a program that uses a for loop to calculate the sum of numbers from 1 to 100.
num =100
x = 0
for i in range(1,num+1):
  x = x + i
print(x)
```

```python
#4.	Write a program to find the largest of three numbers using if, elif, and else.
a = int(input("ENTER FIRST NUMBER: "))
b = int(input("ENTER SECOND NUMBER: "))
c = int(input("ENTER THIRD NUMBER: "))
if a>b>c:
  print("FIRST NUMBER IS MAXIMUM")
  print("LAST NUMBER IS MINIMUM")
elif a>c>b:
  print("FIRST NUMBER IS MAXIMUM")
  print("SECOND NUMBER IS MINIMUM")
elif b>a>c:
  print("SECOND NUMBER IS MAXIMUM")
  print("THIRD NUMBER IS MINIMUM")
elif b>c>a:
  print("SECOND NUMBER IS MAXIMUM")
  print("FIRST NUMBER IS MINIMUM")
elif c>b>a:
  print("THIRD NUMBER IS MAXIMUM")
  print("FIRST NUMBER IS MINIMUM")
else:
  print("THIRD NUMBER IS MAXIMUM")
  print("SECOND NUMBER IS MINIMUM")
```

```python
#5.	Write a program that counts how many numbers from 1 to 30 are divisible by 3 using a for loop.
x = 0
for i in range(1,31):
  if i%3==0:
    x = x +1
print(x)
```

```python
#6.	Write a while loop program that keeps asking for input until the user enters the word stop.
string = input()
x = string.lower()
while x!="stop":
  string = input()
print("if input stop so loop is stop")
```

```python
#7.	Write a program using if elif else to check if a number is positive, negative, or zero.
num = int(input())
if num > 0:
  print("positive")
elif num<0:
  print("negative")
else:
  print("zero")
```

```python
#8.	Write a for loop to iterate through a string and print each character.
string = input("ENTER A STRING:")
for i in string:
  print(i)
```

```python
#9.	Write a program that prints the factorial of a number using a while loo
num = int(input())
x = 1
while num>0:
  x = x * num
  num = num -1
print(x)
```

```python
#10.	Write a program that breaks the loop when the number 5 is encountered in a list.
x = [3,2,6,7,1,9,5,8,4]
for i in x:
  if i == 5:
    break
  print(i, end=" ")
```

```python
#11.	Write a program to print the multiplication table of a number using a for loop.
num = int(input())
for i in range(1,11):
  mul = num * i
  print(f"{num} * {i}   =   {mul}")
```

```python
#12.	Write a program that uses continue to skip printing the number 7 while printing numbers 1 to 10.
for i in range(1,11):
  if i == 7:
   continue
  print(i, end=" ")
```

```python
#13.	Write a program using nested loops to print a 3 by 3 matrix pattern of numbers.
k = int(input())
for i in range(0,3):
  for j in range(0,3):
    print(k , end=" ")
    k = k + 1
  print()
```

```python
#14.	Write a program that prints only the vowels from a given string using a for loop.
string = input()
vow = 'aeiouAEIOU'
for i in string:
  if i in vow:
   print(i, end=" ")
```

```python
#15.	Write a while loop that prints numbers from 10 down to 1.
num = 10
while num>0:
  print(num, end=" ")
  num = num -1
```

```python
#16.	Write a program that reads marks and prints grade using if elif else conditions.
marks = int(input())
if marks>=75:
  print("GRADE A ")
elif marks>=60:
  print("GRADE B")
else:
  print("GRADE C")
```

```python
#17.	Write a for loop to find the count of vowels in a string.
string = input()
vow = 'aeiouAEIOU'
count = 0
for i in string:
  if i in vow:
    count += 1
print(count)
```

```python
#18.	Write a program using a while loop that stops when a number greater than 50 is entered.
x = int(input())
while x<=50:
  x = int(input())
else:
  print("ENTER NUM IS MORETHAN 50")
```

```python
#19.	Write a program that checks if a number is prime using a for loop.
x = int(input())
if x<=1:
  print("not prime")
else:
  k = "prime"
  for i in range(2,x):
      if x % i==0:
         k = "NOT PRIME"
         break
  print(k)

```

```python
#20.	Write a program using nested if conditions to determine if a year is a leap year.
y = int(input("ENTER A YEAR:"))
if (y%4==0 and y%100!=0) or (y%400==0):
  print("LEAP YEAR")
else:
  print("NOT LEAP YEAR")

```

```python
#1.	Write a for loop to print all keys of a dictionary.
d1 = {"name":'vihara',"quality":'best',"top":1, "city":'hyd',"year":2020}
for i in d1.keys():
  print(i, end=" , ")
```

```python
#2.	Write a for loop to print all values of a dictionary.
d1 = {"name":'vihara',"quality":'best',"top":1, "city":'hyd',"year":2020}
for i in d1.values():
  print(i, end=" , ")
```

```python
#3.	Write a for loop to print each key and its corresponding value.
d1 = {"name":'vihara',"quality":'best',"top":1, "city":'hyd',"year":2020}
for i in d1.items():
  print(i, end=" , ")
```

```python
#4.	Using a while loop, iterate through dictionary keys after converting them into a list.
d1 = {"name":'vihara',"quality":'best',"top":1, "city":'hyd',"year":2020}
a =list(d1.keys())
x =len(a)-1
i=0
while i<=x:
  print(a[i], end=" , ")
  i +=1
```

```python
#5.	Write a program to count how many values in a dictionary are integers using a for loop.
d1 = {"name":'vihara',"quality":'best',"top":1, "city":'hyd',"year":2020}
count =0
for i in d1.values():
  if type(i) == int:
    count +=1
print(count)
```

```python
#6.	Write a for loop to find the sum of all numeric values in a dictionary.
d1 = {"name":'vihara',"quality":'best',"top":1, "city":'hyd',"year":2020}
sum =0
for i in d1.values():
  if type(i) == int:
    sum += i
print("sum of numeric is :", sum)
```

```python
#7.	Write a while loop to print dictionary items until all keys are printed.
d1 = {"name":'vihara',"quality":'best',"top":1, "city":'hyd',"year":2020}
l1 = list(d1.keys())
l2 = list (d1.values())
b = len(d1)-1
i=0
while i<=b:
  print(f"{l1[i]} :  {l2[i]}", end=" , ")
  i +=1
```

```python
#8.	Write a program to reverse iterate over dictionary keys using a while loop.
d1 = {"name":'vihara',"quality":'best',"top":1, "city":'hyd',"year":2020}
l1 = list(d1.keys())
l2 = list (d1.values())
b = len(d1)-1
i=b
while i>=0:
  print(f"{l1[i]} : {l2[i]}", end=" , ")
  i -=1
```

```python
#9.	Write a for loop to print keys whose values are greater than a given number.
sal = {"raja":20000,"rani":30000,"roopa":25000, "venkat":50000,"reddy":10000, "yadav":42000}
sv = int(input("enter a value: "))
for i in sal.keys():
   if sal[i] > sv:
     print(f"{i} : {sal[i]}")
```

```python
#10.	Write a program to create a new dictionary with only those key value pairs where the value is even.
sal = {"raja":22,"rani":31,"roopa":25, "venkat":54,"reddy":12, "yadav":42,"bharath":37,"bharathi":40,"sujatha":28}
for i in sal.keys():
   if sal[i]%2==0:
     print(f"{i} : {sal[i]}")
```

```python
#11.	Write a for loop to check if a particular key exists in a dictionary.
sal = {"raja":20000,"rani":30000,"roopa":25000, "venkat":50000,"reddy":10000, "yadav":42000}
sv = input("enter a key: ")
for i in sal.keys():
   if sv == i:
      print("EXIST")
      break
else:
 print("NOT EXIST")
```

```python
#12.	Write a program using while loop to print alternate dictionary keys.
sal = {"raja":22,"rani":31,"roopa":25, "venkat":54,"reddy":12, "yadav":42,"bharath":37,"bharathi":40,"sujatha":28}
l1 = list(sal.keys())
x =len(l1)-1
i=0
while i<=x:
  if i%2==0:
    print(l1[i])
  i +=1
```

```python
#13.	Write a program to print the length of each value if the value is a string, using a for loop.
d1 = {"name":'vihara',"quality":'best',"top":1, "city":'hyderabad',"year":2020}
for i in d1.values():
  if type(i)==str:
    print(len(i))
```

```python
#14.	Write a program to print keys that start with a specific letter.
d1 = {"name":'vihara',"quality":'best',"top":1, "city":'hyderabad',"year":2020,"native":'kakinada'}
sp =input("enter specific letter:")
x = sp.lower()
for i in d1.keys():
  if i[0]==x:
    print(i)

```

```python
#15.	Write a while loop program to iterate through dictionary items by manually tracking an index.
d1 = {"name":'vihara',"quality":'best',"top":1, "city":'hyd',"year":2020}
l1= list(d1.items())
x = len(l1) - 1
i=0
while i <= x:
  a=l1[i][0]
  b=l1[i][1]
  print(f"{a} : {b}")
  i += 1
```

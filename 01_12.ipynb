```python
string = input("Enter a string: ")

vowels = "aeiouAEIOU"
vowel_count = 0
consonant_count = 0

for char in string:
    if char.isalpha():   # check only alphabetic characters
        if char in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

print("Vowels:", vowel_count)
print("Consonants:", consonant_count)

```

```python
#Write a program to check if a given string is a palindrome or not.
num = (input("ENTER NUMBER: "))
pal =num[: :-1]
if pal == num:
  print("PALINDROME")
else:
  print("NOT PALINDROME")
```

```python
#Write a Python program to check whether a character is alphabet, digit, or special character.
char = input("enter char:")
if char.isdigit():
  print("GIVEN CHARACTER IS DIGIT")
elif char.isalpha():
  print("GIVEN CHARACTER IS ALPHABET")
else:
  print("GIVEN CHARACTER IS SPECIAL CHAR")
```

```python
# .	Write a Python program to find the maximum and minimum of three numbers using conditional statements
a = int(input("ENTER FIRST NUMBER: "))
b = int(input("ENTER SECOND NUMBER: "))
c = int(input("ENTER THIRD NUMBER: "))
if a>b>c:
  print("FIRST NUMBER IS MAXIMUM")
  print("LAST NUMBER IS MINIMUM")
elif a>c>b:
  print("FIRST NUMBER IS MAXIMUM")
  print("SECOND NUMBER IS MINIMUM")
elif b>a>c:
  print("SECOND NUMBER IS MAXIMUM")
  print("THIRD NUMBER IS MINIMUM")
elif b>c>a:
  print("SECOND NUMBER IS MAXIMUM")
  print("FIRST NUMBER IS MINIMUM")
elif c>b>a:
  print("THIRD NUMBER IS MAXIMUM")
  print("FIRST NUMBER IS MINIMUM")
else:
  print("THIRD NUMBER IS MAXIMUM")
  print("SECOND NUMBER IS MINIMUM")
```

```python
#	Write a Python program to replace all spaces in a string with '-' symbol.
string =input("enter a string:")
space = " "
for i in string:
  if i == " ":
    space += "-"
  else:
    space += i
print(space)
```

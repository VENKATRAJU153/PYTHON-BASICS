```python
#1.	Write a for loop program to print all elements of a tuple.
t1 = ('vihara', 'is','best','instiute','in', 'hyd')
for i in t1:
  print(i)
```

```python
#2.	Write a while loop program to print all elements of a tuple using indexes.
t1 = ('vihara', 'is','best','instiute','in', 'hyd')
x = len(t1)-1
i=0
while i<=x:
  print(t1[i], end=" ")
  i +=1
```

```python
#3.	Write a for loop program to count how many even numbers are present in a tuple.
t1 = (22,35,12,29,67,44,23,19,76,56,18)
count= 0
for i in t1:
  if i%2==0:
    count +=1
print(f"total even no. in tuple is: {count}")
```

```python
#4.	Write a while loop program to find the largest value in a tuple.
t1 = (22,35,12,29,67,44,23,19,76,56,18)
x = len(t1)-1
i = 0
l =0
while i<=x:
  if t1[i]>l:
    l=t1[i]
  i +=1
print("largest value in a tuple is:", l)
```

```python
#5.	Write a for loop program to print the index and value of each tuple element.
t1 = ('vihara', 'is','best','instiute','in', 'hyd')
for i in t1:
  print(t1.index(i), " ", i)
```

```python
#6.	Write a while loop program to reverse a tuple manually without using slicing.
t1 = ('vihara', 'is','best','instiute','in', 'hyd')
x = len(t1)-1
i=x
while i>=0:
  print(t1[i], end=" ")
  i -=1
```

```python
#7.	Write a for loop program to create a new tuple containing only odd numbers from an existing tuple.
t1 = (22,35,12,29,67,44,23,19,76,56,18)
l1 = []
count= 0
for i in t1:
  if i%2!=0:
    l1.append(i)
new_t2 =tuple(l1)
print(f" odd no. in new tuple is: {new_t2}")
```

```python
#8.	Write a while loop program that prints elements until a specific value is found, then stops.
t1 = (22,35,12,29,67,44,23,19,76,56,18)
sv = int(input())
x = len(t1)-1
i = 0

while i<=x:
  if t1[i]==sv:
    break
  print(t1[i], end =" ")
  i +=1
```

```python
#9.	Write a for loop program to calculate the sum of all numbers inside a tuple.
t1 = (22,35,12,29,67,44,23,19,76,56,18)
count =0
for i in t1:
  count += i
print(f"sum of all numbers inside a tuple : {count}")
```

```python
#10.	Write a while loop program to count how many times a particular element occurs in a tuple.
t1 = (22,35,12,29,67,44,67,19,76,67,18)
sv = int(input())
x = len(t1)-1
i = 0
count = 0
while i<=x:
  if t1[i]==sv:
    count +=1
  i +=1
print(count)
```

```python
#11.	Write a for loop program to find how many strings in a tuple have length greater than 5.
t1 = ('vihara', 'is','best','instiute','in', 'hyderabad', 'and','also','bengalore')
count = 0
for i in t1:
  if len(i)>5:
    count +=1
print(count)
```

```python
#12.	Write a while loop program to print alternate elements of a tuple.
t1 = (22,35,12,29,67,44,67,19,76,67,18)
x = len(t1)-1
i = 0
while i<=x:
  if i%2==0:
    print(t1[i], end=" ")
  i +=1
```

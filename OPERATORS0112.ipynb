```python
#8.	Write a program to check whether a given number is both divisible by 3 and 5.
x =int(input())
if x%3==0 and x%5==0:
  print(" Given num is divisible both 3 and 5")
else:
  print(" Given num is not divisible both 3 and 5")
```

```python
#9.	Write a program that accepts three numbers and prints the largest using comparison and logical operators.
a = int(input("ENTER FIRST NUMBER: "))
b = int(input("ENTER SECOND NUMBER: "))
c = int(input("ENTER THIRD NUMBER: "))
if a>b>c:
  print("FIRST NUMBER IS MAXIMUM")
  print("LAST NUMBER IS MINIMUM")
elif a>c>b:
  print("FIRST NUMBER IS MAXIMUM")
  print("SECOND NUMBER IS MINIMUM")
elif b>a>c:
  print("SECOND NUMBER IS MAXIMUM")
  print("THIRD NUMBER IS MINIMUM")
elif b>c>a:
  print("SECOND NUMBER IS MAXIMUM")
  print("FIRST NUMBER IS MINIMUM")
elif c>b>a:
  print("THIRD NUMBER IS MAXIMUM")
  print("FIRST NUMBER IS MINIMUM")
else:
  print("THIRD NUMBER IS MAXIMUM")
  print("SECOND NUMBER IS MINIMUM")
```

```python
#12.	Write a program to toggle the case of a given alphabet using ASCII operators:
alp = input()
x =ord(alp)
A= ord("A")
Z= ord("Z")
if A<=x<=Z:
  x = x + 32
else:
  x = x -32
y = chr(x)
print(y)
```

```python
''''14.	Write a Python program to perform bitwise operations between two numbers:
a.	AND (&)
b.	OR (|)
'''
x = int(input())
y = int(input())
a = x & y
b = x|y
print("AND RESULT:" , a)
print("OR RESULT:", b)
```

```python
#1.	Access the 3rd element of a list.
lst = list(input())
print(lst[2])
```

```python
#2.	Print the last element without using negative index.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
lst.reverse()
print(lst[0])
```

```python
#3.	Slice a list to get first three items.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]

print(lst[0:3])
```

```python
#4.	Slice to get every second element.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
print(lst[: : 2])
```

```python
#5.	Reverse a list using slicing.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
print(lst[: : -1])
```

```python
#6.	Create a shallow copy of a list.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
lst2 = lst.copy()
print(lst2)
```

```python
#7.	Find length of a list.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
print(len(lst))
```

```python
#8.	Find max value in a list.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
lst2 =['22','12','56','34','78','45','37']
print(max(lst))
print(max(lst2))
```

```python
#9.	Find min value in a list.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
lst2 =['22','12','56','34','78','45','37']
print(min(lst))
print(min(lst2))
```

```python
#10.	Remove the last element from a list.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
lst2 =['22','12','56','34','78','45','37']
lst.pop()
lst2.pop()
print(lst)
print(lst2)
```

```python
#11.	Add an item to end of list.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
lst.append("institute")
print(lst)
```

```python
#12.	Insert element at index 1.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
lst.insert(1,"institute")
print(lst)
```

```python
#13.	Convert string to list of characters.
string = input()
x = list(string)
print(x)
```

```python
#14.	Convert string "a b c" to ['a','b','c'].
string = "a b c"
string = string.replace(" ","")
x = list(string)
print(x)
```

```python
#15.	Concatenate two lists.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
lst2 =['22','12','56','34','78','45','37']
lst3 = lst + lst2
lst.extend(lst2)
print(lst3)
print(lst)

```

```python
#16.	Check if value exists in list.
lst2 =['22','12','56','34','78','45','37']
x = (input())
if x in lst2:
  print("EXIST")
else:
  print("NOT EXIST")
```

```python
#17.	Create a list from 1-5 using list constructor.
lst = list(input())
lst2 = list(range(1,6))
print(lst)
print(lst2)
```

```python
#18.	Remove an item by value.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
lst.remove("hyd")
print(lst)
```

```python
#19.	Clear all elements of a list.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
lst.clear()
print(lst)
```

```python
#20.	Sort a list without modifying original list.
lst = ["apple","banana","vihara", "tech","best","one","hyd"]
lst2= list(lst)
print(lst)
print(sorted(lst2))

```

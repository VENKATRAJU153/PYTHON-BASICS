```python
#1.	Take a string from user and print its length.
string = input("enter the string:")
x = len(string)
print(x)
```

```python
#2.	Input a string and print first and last character.
string = input("enter the string:")
x = string[0]
y = string[-1]
print(x , y)
```

```python
#3.	Input a string and print character at index 2 and −2.
string = input("enter the string:")
x = string[2]
y = string[-2]
print(x , y)
```

```python
#4.	Input a string and print the middle character(s).
string = input("enter the string:")
x = len(string)//2
print(string[x])
```

```python
#5.	Reverse a string using slicing
string = input("enter the string:")
x = string[: : -1]
print(x)
```

```python
#6.	Extract first 5 characters and last 5 characters.
string = input("enter the string:")
x = len(string)
if x>=10:
  a = string[: 5]
  b = string[-5 :]
  print(a)
  print(b)
else:
  print("string length is too less")
```

```python
#7.	Print every second character from a string.
string = input("enter the string:")
x = len(string)
if x>=2:
  a = string[: : 2]
  print(a)
else:
  print("string length is too less")
```

```python
#8.	Print characters from 2nd index to 7th index.
string = input("enter the string:")
x = len(string)
if x>=7:
  a = string[2 : 8]
  print(a)
else:
  print("string length is too less")
```

```python
#9.	Remove leading and trailing spaces from input string.
string = input("enter the string:")
print(string.strip(" "))
```

```python
#10.	Convert string to uppercase and lowercase.
string = input("enter the string:")
x = string.upper()
y = string.lower()
print(x)
print(y)
```

```python
#11.	Replace all spaces in a string with -.
string = input("enter the string:")
x = string.replace(" ", "-")
print(x)
```

```python
#12.	Check if a string starts with "A" or ends with "Z".
string = input("enter the string:")
x = string[0]
y = string[-1]
if x =="A" or y=="Z":
 print("TRUE")
else:
  print("FALSE")
```

```python
#13.	Count occurrences of a character in string using count().
string = input("enter the string:")
ch = input("enter the characher:")
x  = string.count(ch)
print(x)
```

```python
#14.	Find position of substring in given string.
string = input("enter the string:")
ch = input("enter the substing:")
x  = string.find(ch)
print(x)
```

```python
#15.	Split a string into words and print each word separately (no loop — use index access methods).
string = input("enter the string:")
x = string.split(" ")
y = len(x)
print(y)
print(x)
print(x[0])
print(x[1])
print(x[2])
print(x[3])
```

```python
#16.	Check if input is numeric or alphabetic.
x = input("enter the string:")
if x.isalpha():
  print("Given string is all alphabet")
elif x.isnumeric():
  print("Given string is all numeric")
else:
  print("Given string is mixed")
```

```python
#17.	Swap first and last characters of string using slicing.
string = input("enter the string:")
x = string[0]
y = string[-1]
x, y = y, x
z = x + string[1:-1] + y
print(z)

```

```python
#18.	Check if two strings are equal (case-insensitive).
string1 =input("enter first string:")
string2 = input("enter second string:")
x = string1.upper()
y = string2.upper()
if x == y:
  print("both are equal")
else:
  print("both are not equal")
```

```python
#19.	Create a substring by skipping characters: s[1:len(s):2].
s = input("enter the string:")
t = s[1:len(s):2]
print(t)
```

```python
#20.	Join a list of strings into one string separated by commas.
l =["vihara","is","my","institue"]
print(f"{l[0]}, {l[1]}, {l[2]}, {l[3]}")
m = ","
n=m.join(l)
print(n)
```

```python
#1.	Write a program using a for loop to print each character of a string on a new line.
string = input()
for i in string:
  print(i)
```

```python
#2.	Write a program using a while loop to print each character of a string with its index number.
string = input()
x = len(string)
i=0
while i < x:
  print(string[i], i)
  i = i +1
```

```python
#3.	Write a for loop program to count how many vowels are in a given string.
string= input()
vow ='aeiouAEIOU'
count =0
for i in string:
  if i in vow:
    count +=1
print(count)
```

```python
#4.	Write a while loop program to reverse a string manually.
string = input()
i = len(string)-1
while i >=0:
  print(string[i], end="")
  i = i - 1
```

```python
#5.	Write a for loop program that prints characters of a string except spaces.
string = input()
x = " "
for i in string:
  if i != x:
    print(i, end="")
```

```python
#6.	Write a program using a while loop to count how many times a specific character appears in a string.
string = input("enter a string :")
ch = input("enter which character to count:")
count = 0
x = len(string)
i = 0
while i < x:
  if ch == string[i]:
    count +=1
  i +=1
print(count)
```

```python
#7.	Write a for loop to print only uppercase characters in a string.
string = input("enter a string:")
upp = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
for i in string:
  if i in upp:
    print(i, end=" ")
```

```python
#8.	Write a while loop program that stops printing characters when it encounters the letter a.
st= input("enter a string :")
string = st.lower()
ch = "a"
x = len(string)
i = 0
while i < x:
  if ch == string[i]:
    break
  print(string[i], end=" ")
  i +=1
```

```python
#9.	Write a program using a for loop to check how many digits exist inside a mixed string.
string = input("ENTER A STRING: ")
d = 0
for i in string:
  if i.isnumeric():
    d +=1
print(d)
```

```python
#10.	Write a while loop program to print alternate characters of a string.
string = input()
x = len(string)
i=0
while i<=x-1:
  if i%2==0:
   print(string[i], end="")
  i +=1
```

```python
#11.	Write a for loop that prints characters of a string along with their ASCII values.
string = input()
for i in string:
  print(i," ", ord(i))
```

```python
#12.	Write a program using a while loop to copy characters of one string into another new string.
st1 = input("ENTER STRING ST1: ")
st2 = ""
x = len(st1)
i = 0
while i < x:
  st2 += st1[i]
  i +=1
print("NEW STRING IS ST2:", st2)
```

```python
from ast import Index
#13.	Write a for loop to print characters of a string in reverse order without slicing.
string = input()
x = len(string)-1
for i in range(x,-1,-1):
  print(string[i], end="")
```

```python
#14.	Write a while loop program that prints the first half of a string.
string = input()
x =len(string)//2
print(f"first half vaue is:{x}")
i=0
while i<=x:
  print(string[i], end="")
  i +=1
```

```python
#15.	Write a for loop program to count how many alphabetic characters, digits, and special characters are in a string.
string = input()
d = 0
a = 0
sc = 0
for i in string:
  if i.isnumeric():
    d +=1
  elif i.isalpha():
    a +=1
  else:
    sc +=1
print(f"NO OF DIGITS IS:{d}")
print(f"NO OF ALPHABETS IS:{a}")
print(f"NO OF SPECIAL CHAR IS:{sc}")


```

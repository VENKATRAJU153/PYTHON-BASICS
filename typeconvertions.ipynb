```python
# INT TO FLOAT CONVERTION
a=25
print(a)
print(type(a))
print(id(a))
b=float(a)
print(b)
print(type(b))
print(id(b))
```

```python
# INT TO STRING CONVERTION
k=880
print(k)
print(type(k))
print(id(k))
l=str(k)
print(l)
print(type(l))
print(id(l))
```

```python

# MY PRACTICE YEAR AND SALARY
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression



f= {'YEAR':[2012,2015,2017,2019,2020,2021,2022,2023],
    'SALARY':[25000,35000,45000,60000,45000,35000,80000,50000]}
df = pd.DataFrame(f)
df
```

```python
x =df[['YEAR']]
y = df[['SALARY']]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
trained_data = pd.DataFrame({'X-TRAIN':x_train,'Y-TRAIN':y_train})
trained_data
```

```python
tested_data =pd.DataFrame({'X-TEST':x_test,'Y-TEST':y_test})
tested_data
```

```python

x_train.shape
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
# M VALUE
reg.coef_
```

```python
# c value
reg.intercept_
```

```python
trained_data
```

```python
raj_answer = reg.predict(x_train)
raj_answer
```

```python
trained_data['raj_answer'] =raj_answer
trained_data
```

```python
from sklearn.metrics import r2_score
r2_score(y_train,raj_answer)*100
```
